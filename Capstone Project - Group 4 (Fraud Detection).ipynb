<b>Grid-Search on Random Forest</b>

Using cross-validation

```python
#importing libraries
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import make_scorer
VERBOSE_OUTPUT_LEVEL = 3  #To visualize Grid search results
refit = True   #To fit the best model of grid search to training data
cv = 5   #cv - 5 fold cross validation during training
```

```python
criterion = ['gini','entropy']
max_depth = [2,5]
min_samples_split = [2, 5]
param_grid = {'criterion': criterion, 'max_depth': max_depth, 'min_samples_split': min_samples_split}
```

```python
clf = RandomForestClassifier(random_state=13, max_features='auto')
gs_cv = GridSearchCV(clf, param_grid, cv=cv, scoring='f1_micro', refit=refit, verbose=VERBOSE_OUTPUT_LEVEL)
```

```python
gs_clf = gs_cv.fit(X_train,y_train)
```

```python
best_clf = gs_clf.best_estimator_
clf_params = gs_clf.best_params_
clf_score = gs_clf.best_score_
print(best_clf)
print(f"best score {clf_score}, best parameters {clf_params}")
```

```python
# score
accuracy = round(cross_val_score(best_clf, X, y, scoring=make_scorer(accuracy_score), cv = cv).mean(),4)
balanced_accuracy = round(cross_val_score(best_clf, X, y, scoring=make_scorer(balanced_accuracy_score), cv = cv).mean(),4)
f1_score = round(cross_val_score(best_clf, X, y, scoring=make_scorer(f1_score), cv = cv).mean(),4)
print(f"Accuracy on Training Set : {accuracy}")
print(f"Balanced Accuracy on Training Set : {balanced_accuracy}")
print(f"F1 Score on Training Set : {f1_score}")
```

```python
dt_yhat = best_clf.predict(X_test)
dt_yhat
```
